# Image to sketch convertion using python

```python
import matplotlib.pyplot as plt
import pandas as pd
import cv2
```

```python
img = cv2.imread('C:\\Users\\ADMIN\\jupyterFiles\\pencil-sketch\\sample.jpg',1)
img_resized = cv2.resize(img, (1920, 1080), interpolation=cv2.INTER_LINEAR)
```

### Show image
Figure size taken is 960x540 which is half of its original dimension

```python
plt.subplots(figsize=(9.60,5.40))
plt.imshow(cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB))
plt.show()
```

### 1. RGB to Grayscale
We need to read the image in RBG format and then convert it to a grayscale image.
This will turn an image into a classic black and white photo.

```python
plt.subplots(figsize=(9.60,5.40))
image_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_resized = cv2.resize(image_grey, (1920, 1080), interpolation=cv2.INTER_LINEAR)
plt.imshow(cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB))
plt.show()
```

### 2. Grayscale to Negative
inverting the grayscale image also called negative image, this will be our inverted grayscale image.

```python
plt.subplots(figsize=(9.60,5.40))
image_invert = cv2.bitwise_not(image_grey)
img_resized = cv2.resize(image_invert, (1920, 1080), interpolation=cv2.INTER_LINEAR)
plt.imshow(cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB))
plt.show()
```

### 3. Negative to Blur
we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.



```python
blur = cv2.GaussianBlur(image_invert, (21, 21), 0)
```

```python
blur_invert = 255-blur
```

```python
plt.subplots(figsize=(9.60,5.40))
final_sketch = cv2.divide(image_grey, blur_invert, scale=256.0)
img_resized = cv2.resize(final_sketch, (1920, 1080), interpolation=cv2.INTER_LINEAR)
plt.imshow(cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB))
plt.show()
```
